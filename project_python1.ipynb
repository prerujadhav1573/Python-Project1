```python
# Simple General Store Management System

print("Welcome to Sakhi General Store")

products = {
    1: {"name": "Pen", "price": 10},
    2: {"name": "Pencil", "price": 5},
    3: {"name": "Notebook", "price": 50},
    4: {"name": "Book", "price": 120},
    5: {"name": "Eraser", "price": 3}
}

cart = []  

while True:
    print("\nAvailable Products:")
    for pid in products:
        print(pid, ".", products[pid]["name"], "- ₹", products[pid]["price"])


    product_id = int(input("\nEnter product ID to buy: "))

    if product_id in products:
        qty = int(input("Enter quantity: "))
        discount = int(input("Enter discount (5 or 10): "))

        total = products[product_id]["price"] * qty
        discount_amt = (total * discount) / 100
        final_price = total - discount_amt

        cart.append({
            "item": products[product_id]["name"],
            "qty": qty,
            "discount": discount,
            "total": final_price
        })

        print(products[product_id]["name"], "added to cart.")
    else:
        print("Invalid product ID!")

    more = input("Do you want to buy more items? (yes/no): ").lower()
    if more != "yes":
        break


print("\n------- BILL -------")
grand_total = 0
for item in cart:
    print(item["item"], "(x", item["qty"], ")", "- ₹", item["total"])
    grand_total += item["total"]

print("--------------------")
print("Total Amount to Pay: ₹", grand_total)


print("\nSelect Payment Method:")
print("1. Cash")
print("2. Online")
print("3. Credit Card")

choice = int(input("Enter your choice: "))
if choice == 1:
    method = "Cash"
elif choice == 2:
    method = "Online"
elif choice == 3:
    method = "Credit Card"
else:
    method = "Invalid"

print("\nPayment Method:", method)
print("Payment Successful!")
print("Thank you for shopping at Sakhi General Store! Visit again.")
```
